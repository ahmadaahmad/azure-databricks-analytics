```python
from datetime import datetime

storage_account_datalake = "salesdatalakedemo"
container_datalake = "salescontainer"
today = datetime.now()

# input path for stg parquet data
input_path = f"abfss://{container_datalake}@{storage_account_datalake}.dfs.core.windows.net/stg/{today:%Y/%m/%d}"

# Output path for processed parquet data
output_path = f"abfss://{container_datalake}@{storage_account_datalake}.dfs.core.windows.net/processed/{today:%Y/%m/%d}"

df_stg = spark.read.parquet(input_path)
df_stg.createOrReplaceTempView("df_stg")

# Here, I read cleaned data from temp view or from raw and transform again:
df_clean = spark.sql("SELECT *, getdate() as processed_on FROM df_stg")

# Write as parquet (columnar optimized)
df_clean.write.mode("overwrite").parquet(output_path)

print(f"Cleaned data saved as parquet at: {output_path}")
```
