```python
# Parameters - replace with your values or secrets retrieval
#
storage_account_blob = "salesblobstorage"
container_blob = "salesdatacontainer"
file_name = "Sample - Superstore.csv"

storage_account_datalake = "salesdatalakedemo"
container_datalake = "salescontainer"

client_id = dbutils.secrets.get(scope="salesdbscope", key="client-id")
tenant_id = dbutils.secrets.get(scope="salesdbscope", key="tenant-id")
client_secret = dbutils.secrets.get(scope="salesdbscope", key="secret-value")

# Set configs for Blob Storage OAuth (abfss)
spark.conf.set(
    f"fs.azure.account.auth.type.{storage_account_blob}.dfs.core.windows.net", "OAuth"
)
spark.conf.set(
    f"fs.azure.account.oauth.provider.type.{storage_account_blob}.dfs.core.windows.net",
    "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider"
)
spark.conf.set(
    f"fs.azure.account.oauth2.client.id.{storage_account_blob}.dfs.core.windows.net",
    client_id
)
spark.conf.set(
    f"fs.azure.account.oauth2.client.secret.{storage_account_blob}.dfs.core.windows.net",
    client_secret
)
spark.conf.set(
    f"fs.azure.account.oauth2.client.endpoint.{storage_account_blob}.dfs.core.windows.net",
    f"https://login.microsoftonline.com/{tenant_id}/oauth2/token"
)

spark.conf.set(
f"fs.azure.account.key.{storage_account_blob}.blob.core.windows.net",
"<BLOB STORAGE ACCESS KEY>"
)

# Set configs for Data Lake Gen2 OAuth (abfss)
spark.conf.set(
    f"fs.azure.account.auth.type.{storage_account_datalake}.dfs.core.windows.net", "OAuth"
)
spark.conf.set(
    f"fs.azure.account.oauth.provider.type.{storage_account_datalake}.dfs.core.windows.net",
    "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider"
)
spark.conf.set(
    f"fs.azure.account.oauth2.client.id.{storage_account_datalake}.dfs.core.windows.net",
    client_id
)
spark.conf.set(
    f"fs.azure.account.oauth2.client.secret.{storage_account_datalake}.dfs.core.windows.net",
    client_secret
)
spark.conf.set(
    f"fs.azure.account.oauth2.client.endpoint.{storage_account_datalake}.dfs.core.windows.net",
    f"https://login.microsoftonline.com/{tenant_id}/oauth2/token"
)
```
