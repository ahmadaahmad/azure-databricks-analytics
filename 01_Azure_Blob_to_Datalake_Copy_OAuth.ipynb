```python
# Compose input path for Blob Storage (abfss)
input_path = (
    f"wasbs://{container_blob}@{storage_account_blob}.blob.core.windows.net/{file_name}"
)
print("Reading from Blob Storage URL:")
print(input_path)

# Read CSV from Blob Storage
df_raw = (
    spark.read.option("header", "true").csv(input_path)
)

# Show sample to verify
display(df_raw)

# Prepare dynamic folder structure based on current date for Data Lake Gen2
today = datetime.now()
year = today.strftime("%Y")
month = today.strftime("%m")
day = today.strftime("%d")

# Compose output folder path on Data Lake Gen2 with subfolders raw/year/month/day/
output_folder = (
    f"abfss://{container_datalake}@{storage_account_datalake}.dfs.core.windows.net/raw/{year}/{month}/{day}"
)

# Write DataFrame as CSV to Data Lake Gen2 (overwrite or append depending on need)
df_raw.write.mode("overwrite").option("header", "true").csv(output_folder)

print(f"File copied from Blob Storage to Data Lake Gen2 at: {output_folder}")
```
